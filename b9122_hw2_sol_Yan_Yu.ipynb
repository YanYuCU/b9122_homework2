```python
from bs4 import BeautifulSoup
import urllib.request
```

## Q1.1

```python
seed_url = "https://www.federalreserve.gov/newsevents/pressreleases.htm"

urls = [seed_url]      #queue of urls to crawl
seen = set([seed_url]) #stack of urls seen so far
rst  = []              

maxNumUrl = 10;      #set the maximum number of urls to collect
print("Starting with url="+str(urls))


while len(urls) > 0 and len(rst) < maxNumUrl:
    # DEQUEUE A URL FROM urls AND TRY TO OPEN AND READ IT
    try:
        curr_url=urls.pop(0)
        print("num. of URLs in stack: %d " % len(urls))
        print("Trying to access= "+curr_url)
        req = urllib.request.Request(curr_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()

    except Exception as ex:
        print("Unable to access= "+curr_url)
        print(ex)
        continue

    
    soup = BeautifulSoup(webpage)  
    for tag in soup.find_all('a', href = True): #find tags with links
        childUrl = tag['href'] 
        newUrl = urllib.parse.urljoin(seed_url, childUrl)
        print("seed_url= " + seed_url)
        print("original childurl= " + childUrl)
        print("new childurl= " + newUrl)
        print("seed_url in the new childUrl= " + str("https://www.federalreserve.gov/newsevents/pressreleases" in newUrl))
        print("Have we seen this childUrl= " + str(newUrl in seen))
        if "https://www.federalreserve.gov/newsevents/pressreleases" in newUrl and newUrl not in seen:
            urls.append(newUrl)
            seen.add(newUrl)
        else:
            print("######")
            continue
        
        
        req = urllib.request.Request(newUrl,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        if 'covid' in BeautifulSoup(webpage).get_text().lower():
            rst.append(newUrl)
        if len(rst) == maxNumUrl:
            break
```

```python
print("List of URLs with word 'covid':")
for i in range(len(rst)):
    print(str(i+1)+'.\tURL: ', rst[i])
```

## Q1.2

```python
seed_url = "https://www.sec.gov/news/pressreleases"

urls = [seed_url]      #queue of urls to crawl
seen = set([seed_url]) #stack of urls seen so far
rst  = []              

maxNumUrl = 20;      #set the maximum number of urls to collect
print("Starting with url="+str(urls))


while len(urls) > 0 and len(rst) < maxNumUrl:
    # DEQUEUE A URL FROM urls AND TRY TO OPEN AND READ IT
    try:
        curr_url=urls.pop(0)
        print("num. of URLs in stack: %d " % len(urls))
        print("Trying to access= "+curr_url)
        req = urllib.request.Request(curr_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()

    except Exception as ex:
        print("Unable to access= "+curr_url)
        print(ex)
        continue

    
    soup = BeautifulSoup(webpage)  
    for tag in soup.find_all('a', href = True): #find tags with links
        childUrl = tag['href'] 
        newUrl = urllib.parse.urljoin(seed_url, childUrl)
        print("seed_url= " + seed_url)
        print("original childurl= " + childUrl)
        print("new childurl= " + newUrl)
        print("seed_url in the new childUrl= " + str('https://www.sec.gov/news/press-release' in newUrl))
        print("Have we seen this childUrl= " + str(newUrl in seen))
        if 'https://www.sec.gov/news/press-release' in newUrl and newUrl not in seen:
            urls.append(newUrl)
            seen.add(newUrl)
        else:
            print("######")
            continue
            
        if 'charges' in tag.text.lower():
            rst.append([newUrl, tag.text])
        if len(rst) == maxNumUrl:
            break
```

```python
print("List of URLs with word 'charges':")
for i in range(len(rst)):
    print(str(i+1)+'.\tURL: ', rst[i][0])
    print(str(i+1)+'.\tText: ', rst[i][1])
    print()
```
